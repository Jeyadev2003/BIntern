```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L        
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_director)
```

```python
movies_df
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies_df
```

```python
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
new_df=movies_df[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)     
```

```python
new_df['tags']=new_df['tags'].apply(stem) 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse= True, key=lambda x:x[1])[1:6]
```

```python
from fuzzywuzzy import fuzz, process

def recommend(movie):
    movie_lower = movie.lower()
    if movie_lower not in new_df['title'].str.lower().values:
        closest_match, _ = process.extractOne(movie_lower, new_df['title'].str.lower().values)

        print()
        user_response = input(f"'{movie}' Not Found. Did you mean '{closest_match.capitalize()}'? (yes/no): ").strip().lower()

        if user_response == "yes":
            movie = closest_match.capitalize()
        else:
            return []

    movie_index = new_df[new_df['title'].str.lower() == movie.lower()].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]
    recommended_movies = [new_df.iloc[i[0]].title for i in movies_list]
    return recommended_movies

def display_in_box(movie_list):
    max_length = max(len(movie) for movie in movie_list)
    horizontal_line = "+" + "-" * (max_length + 2) + "+"

    print(horizontal_line)
    for movie in movie_list:
        print(f"| {movie.ljust(max_length)} |")
    print(horizontal_line)

user_input = input("Enter the name of the movie: ")

recommended_movies = recommend(user_input)

if recommended_movies:
    print("\nRecommended movies")
    display_in_box(recommended_movies)
```
